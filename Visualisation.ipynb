```python
df=a.merge(b,left_on='director_id',right_on='id',how='left')
df
```

```python
df.drop(columns=['Unnamed: 0_x','id_x','Unnamed: 0_y','id_y'],inplace=True)
```

```python
df['year'].value_counts()
```

```python
df.sort_values(by='director_name').reset_index(drop=True)
```

# Bar Graph

```python
df.groupby('year')['popularity'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(12,8))
plt.bar(df['year'],df['popularity'],color='r')
plt.title('Popularity of film')
plt.xlabel('Year')
plt.ylabel('Popularity')
plt.grid()
plt.show()
```

* the film raleased in 2014 is most popular
* the popularity of film increases as year goes
* the popularity of the film is maximum at 2010 and 2015


```python
director_revenue = df.groupby('director_name')['revenue'].sum().sort_values(ascending=False)
a=director_revenue.head(10)
plt.figure(figsize=(12, 8))
a.plot(kind='bar', color='skyblue')
plt.title('Total Revenue by Director')
plt.xlabel('Director')
plt.ylabel('Total Revenue')
plt.xticks(rotation=90)
plt.grid()
plt.show()
```

* the graph shows the comparison of total revenue for each director.
* the movies directed by Steven Spielberg have maximum revenue

# Pie Chart

```python
df.groupby('title')['budget'].sum().sort_values(ascending=False).head(5).plot(kind='pie',autopct='%1.2f%%')
plt.title('Budget range of Movie')
plt.show()
```

* this chart shows the top 5 movies with maximum budget
* Pirates of the Caribbean: On Stranger Tides is top movie with maximum budget


```python
df['day'].value_counts().plot(kind='pie',autopct='%1.2f%%')
plt.title('Distribution of movie released  in each day')
```

* This chart shows percentage of movie released in each day
* most of the filims are released in friday

```python
plt.boxplot(df['vote_average'])
plt.title('boxplot of average vote')
plt.show()
```

* there is so many outliers in this data

```python
plt.violinplot(df['vote_average'])
plt.show()
```

* this data is equally distriuted

```python
df['director_name'].value_counts().head(10).plot(kind='bar',color='gold')
plt.xticks(rotation=90)
plt.xlabel('Director Name')
plt.ylabel('Count')
plt.title('Count of Movies Directed')
plt.show()
```

* Bar graph shows no.of movies directed by director
* Steven Spielberg directed more number of filim

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
axes[0].plot(df['year'],df['revenue'])
axes[0].set_title(' Total Revenue')
axes[0].set_xlabel('year')
axes[0].set_ylabel('Revenue')
axes[1].plot(df['year'],df['budget'])
axes[1].set_title('total budget')
axes[1].set_xlabel('year')
axes[1].set_ylabel('Budget')
plt.show()
```
